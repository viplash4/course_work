### Методи змінної метрики

Для застосування методу Пірсона потрібно знайти обернену матрицю Гессе для обраної функції,що може викликати певні труднощі.

Методи змінної метрики допомагають апроксимувати матрицю Гессе, використовуючи тільки перші похідні. У таких методах використовуються спряжені напрямки. Перехід з точки $x^{(k)}$ в точку $x^{(k+1)}$ визначається наступним чином:
$$x^{(k+1)} = x^{(k)} - \lambda^{*(k)} A(x^{(k)})\nabla f(x^{(k)}) $$ де $A(x^{(k)})$ - матриця $n$x$n$ яка являэ собою апроксимацію оберненої матриці Гессе.

Для апроксимації матриці, оберненої до матриці Гессе, використовується співвідношення:
$$A^{(k+1)} = A^{(k)} + A_c^{(k)} $$
де $A_c^{(k)}$

### Метод Пірсона
Метод апроксимує матрицю Гессе або обернену до неї, але використовують для цього лише перші похідні.
На кожному кроці $ x^{k+1} = x^k + \lambda_{k}S_{k}  $, де $S^k= - \widetilde{H}_{k} \nabla f(x^k) $ - напрямок пошуку. В цьому алгоритмі обернений гесіан апроксимується матрицею, яка обчислюється на кожному кроці за формулою:
$$ \widetilde{H}_{k=1} = \widetilde{H}_{k} + \frac{((x^{k+1}-x^k)-\widetilde{H}_{k}(\nabla f(x^{k+1})-\nabla f(x^k)))(x^{k+1}-x^k)}{(x^{k+1}-x^k)(\nabla f(x^k)- \nabla f(x^k))}$$
або в іншій модифікації:
$$ \widetilde{H}_{k+1} = \widetilde{H}_{k} + (x^{(k+1)}- x^{(k)}- \widetilde{H}_{k}  \cdot(\nabla f(x^{(k+1)} - \nabla f(x^{k}))) \cdot A_{k}$$, де $$ A_{k} = 
\frac{\widetilde{H}_{k}\cdot(\nabla f(x^{(k+1)}-\nabla f(x^{(k)})^T}{(\nabla f(x^{(k+1)}-\nabla f(x^{(k)})^T\cdot \widetilde{H}_{k} \cdot (\nabla f(x^{(k+1)} - \nabla f(x^{(k)})  } $$

В якості початкової матриці $\widetilde{H}_{0}$ обирається визначена додатньо симетрична матриця (зазвичай одинична). Даний алгоритм Пірсона часто приводить до ситуацій, коли матриця $\widetilde{H}_{k}$ стає погано зумовленою, а саме - вона починає коливатись між додатніми та від'ємними значеннями при цьому визначник матриці близький до нуля. Для того щоб уникнути цієї ситуації, необхідно через кожні $ n $ кроків в якості матриці змінної метрики вибирати матрицю $\widetilde{H}_{0}$

### Чисельне обчислення похідних

Формула правої кінцевої різниці для обчислення першої похідної:

$$f'(x) = \frac{f(x+h) - f(x)}{h}$$

Формула лівої кінцевої різниці для обчислення першої похідної:

$$f'(x) = \frac{f(x) - f(x - h)}{h}$$

Формула для обчислення першої похідної за методом центральних різниць:

$$f'(x) = \frac{f(x+h) - f(x - h)}{2h}$$

### Критерії закічнення

1. $$\begin{cases}
\frac{ | f(x^{k+1}) - f(x^k) | } { | f(x^k) | } <= \varepsilon  \\ 
|f(x^{k+1})-f(x^k))| <= \varepsilon
\end{cases}$$

2. $$ || \nabla f(x^k) || <= \varepsilon $$

### Алгоритм роботи
Спочатку потрібно знайти похідні першого порядку, для заданої функції. Після цього обчислити градієнт за допомогою формули $\nabla f(x) = (\frac{df}{dx} ; \frac{df}{dy})$. За допомогою градієнту можему перейти до знаходження напрямку. $S = - A \nabla f(x)$, паралельно шукаємо довжину кроку $\lambda $ за допомогою  аналітичної формули для обчислення $\lambda_k = - \frac{\nabla^T f(\overline{x}^{(k)})S^{(k)}}{S^{(k)^T}H(\overline{x}^{(k)}S^{(k)}}$. Наступним кроком є обчислення нової точки $x$ та перевірка на критерій закічення.

### Імпорт необхідних бібліотек

```python
import math
import numpy as np
from sympy import Symbol, Derivative, solve, diophantine
```

### Визначення допоміжних функцій



```python
#func 2x1^2+x1x2+x2^2
# x0 = [2,2]
def find_derivaties(func):
  _x1, _x2 = Symbol('x1'), Symbol('x2')
  return Derivative(func,_x1).doit(), Derivative(func,_x2).doit()

def grad(x1,x2,der_x1,der_x2):
  one = der_x1.subs({_x1:x1, _x2:x1})
  two = der_x2.subs({_x1:x1, _x2:x1})
  return np.array([one, two])[np.newaxis].T #without float

def grad_m(x1,x2):
  return np.array([[400 * x1**2 - 400 * x1 * x2 + 2*x1-2, -200*x1**2+200*x2]])
def hess(x1,x2):
  return np.array([
                   [800 * x1 - 400 * x2 + 2,-400*x1],
                   [-400 * x1,200] #write hessian
  ])

def oneDsearch(x1,x2):
  gr = grad_m(x1,x2)
  S = - gr
  num = np.dot(gr, S.T)
  den = np.dot(np.dot(S,hess(x1,x2)), S.T)
  return -(num / den)
```

### Виконання розрахунків

```python
_x1, _x2 = Symbol('x1'), Symbol('x2')
#func = 2*_x1**2+_x1*_x2+_x2**2
func = 100 * (_x1**2 - _x2)**2 + (_x1-1)**2
x0 = [-1.2,0]
a0 = np.matrix('1 0; 0 1')
der_x1, der_x2 = find_derivaties(func)
epsilon = 0.002
finish = 745


grad_fx0 = grad(x0[0],x0[1],der_x1,der_x2)
lambda0 = oneDsearch(x0[0],x0[1])
s0 = np.dot(-a0,grad_fx0)
x1 = np.squeeze(np.asarray(x0+lambda0*s0.T))
grad_fx1 = grad(x1[0],x1[1],der_x1,der_x2)
# 2 iteration
delta_x0 = x1-x0
delta_g0 = grad_fx1 - grad_fx0

_g0 = np.matrix(delta_g0)
_x0 = np.matrix(delta_x0)
a1 = a0 + np.dot(_x0.T, _x0)/np.dot(_x0, _g0) - np.dot(np.dot(np.dot(a0, _g0), _g0.T), a0)/np.dot(np.dot(_g0.T, a0), _g0)
s1 = np.dot(-a1, grad_fx1)
lambda1 = oneDsearch(x1[0],x1[1])
#print(lambda0)
#print(x1)
#print(lambda1)

x2 = np.squeeze(np.asarray(x1 + lambda1*s1.T))
grad_fx2 = grad(x2[0],x2[1],der_x1,der_x2)
h_1 = np.linalg.matrix_power(hess(x0[0],x0[1]), -1)
delta_x1 = x2-x1
delta_g1 = grad_fx2 - grad_fx1

_g1 = np.matrix(delta_g1)
_dx1 = np.matrix(delta_x1)
a2 = a1 + np.dot(_dx1.T, _dx1)/np.dot(_dx1, _g1) - np.dot(np.dot(np.dot(a1, _g1), _g1.T), a1)/np.dot(np.dot(_g1.T, a1), _g1)
#print(grad_fx0,lambda0,s0,x1,grad_fx1,delta_x0,delta_g0)

print(grad_fx2)
print(grad_m(x2[0],x2[1]))
print(type(x2[0]))
```

```python
n = 1 #number of calls

points_log = [x0,x1,x2]
a_log = [a0,a1,a2]
while True:
  grad_fxk = grad(points_log[-1][0],points_log[-1][1],der_x1,der_x2)
  delta_xk = points_log[-1] - points_log[-2]
  delta_gk = grad(points_log[-1][0],points_log[-1][1],der_x1,der_x2) - grad(points_log[-2][0],points_log[-2][1],der_x1,der_x2)
  _gk = np.matrix(delta_gk)
  _xk = np.matrix(delta_xk)
  a_k = a_log[-1] + np.dot(_xk.T, _xk)/np.dot(_xk, _gk) - np.dot(np.dot(np.dot(a_log[-1], _gk), _gk.T), a_log[-1])/np.dot(np.dot(_gk.T, a_log[-1]), _gk)
  sk = np.dot(-a_k, grad_fxk)
  lambdak = oneDsearch(points_log[-1][0],points_log[-1][1])
  #print(lambdak)
  xk = np.squeeze(np.asarray(points_log[-1] + lambdak*sk.T))
  check_final = grad(xk[0],xk[1],der_x1,der_x2)
  #print(check_final)
  #print(type(check_final))
  #print(type(check_final[0][0]))
  points_log.append(xk)
  n += 1
  a_log.append(a_k)
  if np.linalg.norm([float(x[0]) for x in check_final]) <= epsilon:
    print(f'кількість циклів - {n}')
    break
  elif n == finish:
    print(f'кількість циклів - {n}')
    break

print(points_log[-1])
```
